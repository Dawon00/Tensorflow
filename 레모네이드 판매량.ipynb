```python
#라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
#데이터 사용
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
데이터 = pd.read_csv(파일경로)
데이터.head()
```

```python
#종속변수, 독립변수
독립 = 데이터[['온도']]
종속 = 데이터[['판매량']]
print(독립.shape, 종속.shape)#확인해보기
```

```python
#모델을 만들기
X = tf.keras.layers.Input(shape=[1]) #1은 독립변수의 개수
Y = tf.keras.layers.Dense(1)(X) #1은 종속변수의 개수
model = tf.keras.models.Model(X, Y) #모델 완성
model.compile(loss='mse')#모델이 학습할 방법 정의
```

```python
#모델을 학습
model.fit(독립, 종속, epochs=10) #epochs에는 몇번을 학습시킬지 횟수
```
